```python
import cmath
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=1000
```

```python

x = np.random.randint(0, 4, size=N)

def modulator(x):
    if x == 0:
        return complex(-1/cmath.sqrt(2),-1/cmath.sqrt(2))
    if x == 1:
        return complex(-1/cmath.sqrt(2),1/cmath.sqrt(2))
    if x == 2:
        return complex(1/cmath.sqrt(2),-1/cmath.sqrt(2))
    if x == 3:
        return complex(1/cmath.sqrt(2),1/cmath.sqrt(2))
```

```python
symbols = []
for y in x:
    c= modulator(y)
    symbols.append(c)
    

symbols = np.array(symbols)
plt.figure(figsize= (5,5))
plt.plot(symbols.real, symbols.imag, 'bo')
```
