```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 1000
mu = 0
sigma = 1
```

```python
x = np.random.uniform(0,1,M)
w = np.random.normal(mu, sigma, M)
y = 1.2+2.3*x+10*w
```

```python
plt.figure(figsize= (5,5))
plt.plot(x, y, 'o')
plt.xlabel('X')
plt.ylabel('y')
```

```python
plt.subplot(1, 2, 1)
plt.title('Uniform PDF')
_ = plt.hist(x, bins=50, color='b')
plt.subplot(1, 2, 2)
plt.title('Gaussian PDF')
_ = plt.hist(w, bins=50, color='g')
```
