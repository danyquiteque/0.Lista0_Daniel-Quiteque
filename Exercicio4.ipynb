```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
M = 10000
```

```python
# Generating the aleatories distributions
x = np.random.uniform(0,1,M)
y = np.random.uniform(0,2,M)
```

```python
# The z vector will be x+y
z = x+y
```

```python
# Ploting the probability Density Function of z
h = plt.hist(z , bins=100 , density=True, color='b')
```
